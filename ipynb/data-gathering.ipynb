## Reading Dashboard Exported Data

```python
file_name = '../data-raw/fp7-projects.xlsx'
fp7_dashboard = pd.read_excel(file_name, engine='openpyxl')
```

```python
fp7_dashboard = (
    fp7_dashboard
    .clean_names()
    .select_columns([
        "project_nbr",
        "project_acronym",
        "thematic_priority_descr",
        "topic_code",
        "topic_description",
        "participations"])
    .rename(columns={
        "project_nbr": "project_id",
        "thematic_priority_descr": "thematic_priority"})
    .add_column(
        column_name="framework_programme",
        value="FP7")
)
```

```python
file_name = '../data-raw/h2020-projects.xlsx'
h2020_dashboard = pd.read_excel(file_name, engine='openpyxl')
```
